```python
# Declaration data variale
data = [30, 22, 47, 20, 46, 47, 37, 24, 40, 12, 43, 45, 44, 48, 24, 30, 15, 28, 21, 29, 17, 24, 41, 33, 23]
print('Số lượng phần tự trong data list:',len(data))

#Create 1 dim array
arr = np.array(data)
print('\nMảng 1 chiều arr:',arr,'\n')

row = len(data)//5

#Transform 1 dim array to 2 dim array
arr.shape=(int(row),5)
arr
```

```python
#Print the first 5 items in 2 dim array
print('The first 5 items in 2 dim array',arr[0,:])
```

```python
#Print the last 5 items in 2 dim array
print('The last 5 items in 2 dim array',arr[-1,:])
```

```python
# Even (so chan) number condition
def odd_(x):
    return x % 2 != 0
    
print('Số lượng phần tử có giá trị chẵn ',arr.size - np.sum(odd_(arr)))
print('Số lượng phần tử có giá trị lẻ ',np.sum(odd_(arr)))
```

```python
# Declaration arr_2d
arr_2d= arr    

#Replace odd numbers = 1
arr_2d[odd_(arr_2d)] =1

#Replace even numbers = 0
arr_2d[~odd_(arr_2d)] =0
arr_2d
```

```python
#2.1. Load data
with open ('data/Game_of_Thrones_arya_data.csv','r',encoding='utf-8') as f:
    content = f.read()
print(content)
```

```python
#2.2. Cleaning data

content = content.replace('\n','')
content = content.replace('\r','')
content = content.replace(',','')
content = content.replace('?','')
content = content.replace('!','')
content = content.replace('.','')
content = content.replace('\'','')
content = content.replace('\"','')
```

```python
my_stopwords = list(STOPWORDS)
```

```python
#2.3. Create Wordcloud
wc=WordCloud(background_color='white',stopwords=my_stopwords)
wc.generate(content)
```

```python
#2.4. Show Wordcloud
plt.figure(figsize=(20, 6)) 
plt.imshow(wc)
plt.axis('off')
plt.show()
```

```python
#2.5. Create Mask for wordcloud
from PIL import Image
```

```python
img = Image.open('data/Arya.png')
my_mask=np.array(img)
my_mask.shape
```

```python
#2.6 Check image
plt.imshow(my_mask)
plt.axis('off')
plt.show()
```

```python
#2.7 Create wordcloud with mask
wc2= WordCloud(background_color='white',stopwords=my_stopwords,mask=my_mask)
wc2.generate(content)
```

```python
#2.8 Show wordcloud with mask
plt.figure(figsize=(25, 10)) 
plt.imshow(wc2)
plt.axis('off')
plt.show()
```

### 3.5 Vẽ biểu đồ tương quan numlectures và content_duration

```python
# udemy_level= udemy.groupby(by=['level'])['level','num_lectures','content_duration']
# udemy_level

udemy_level = udemy.groupby(by='level')[['num_lectures', 'content_duration']].sum()
print(udemy_level)
```

```python
#subplot 1
# plt.subplot(121)
sns.lmplot(x='num_lectures', y='content_duration', data=udemy, hue="level")
plt.xlabel("Number of Lectures")
plt.ylabel("Content Duration")

#subplot 2
# plt.subplot(122)
sns.jointplot(x='num_lectures', y='content_duration', data=udemy)
plt.xlabel("Number of Lectures")
plt.ylabel("Content Duration")

#Show
plt.subplots_adjust()
plt.show()
```

**Nhận xét chung:**

### 4.2 Xóa các dòng dữ liệu có thuộc tính State bị null

```python
us_unemployment.isnull().any()
```

```python
us_unemployment[us_unemployment['State'].isnull() | us_unemployment['Unemployment_Rate'].isnull()]
```

```python
us_unemployment = us_unemployment.dropna(subset=['State', 'Unemployment_Rate'])
us_unemployment.isnull().any()
```

```python
us_unemployment.info()
```

### 4.3 Nhóm dữ liệu State và tính trung bình (mean) theo Unemployment_Rate

```python
us_unemployment_by_state = us_unemployment.groupby(by=['State'], as_index=False)['Unemployment_Rate'].mean()
us_unemployment_by_state.head()
```

### 4.4 Tạo bản đồ với center là Mỹ (location= [37, -102]) và zoom level (zoom_start=5)

```python
us_map= folium.Map(location = [37,-102], zoom_start = 5)
us_map
```

### 4.5 Tạo choropleth map theo Unemployment_Rate của từng tiểu bang 

```python
# Load "us-states.json"
state_geo = 'data/us-states.json'


#choropleth map
choropleth = folium.Choropleth(
    geo_data=state_geo,
    data=us_unemployment_by_state,
    columns=['State', 'Unemployment_Rate'],
    key_on='feature.id', 
    fill_color='YlOrRd',
    fill_opacity=0.7,
    line_opacity=0.2,
    legend_name='US unemployment rate by state'
).add_to(us_map)  

# Tooltip
choropleth.geojson.add_child(folium.features.GeoJsonTooltip(fields=['name'], labels=False))

#Showmap
us_map
```
